```python
fields_order = [
    'Номер анкеты',
    'Регион',
    'Месяц рождения',
    'Цвет глаз',
    'Цвет волос',
    'Характер',
    'Возможные формы устройства',
    'Группа здоровья',
    'Причина отсутствия родительского попечения матери',
    'Причина отсутствия родительского попечения отца',
    'Братья или сёстры',
    'Дата публикации фотографии',
    'Куда обращаться',
]
```

```python
data_with_fields = []
for item in data:
    fields = {}
    for text_no, text in enumerate(item['texts']):
        if text_no == 2:
            field, value = text.split(' в ', 1)
        elif text_no == 11:
            field, value = text.split(' — ', 1)
        elif text_no == 12:
            field = text
            value = None
        else:
            field, value = text.split(': ', 1)
            field = field.strip()
        if text_no == 2:
            if field == 'Девочка родилась':
                fields['sex'] = 'f'
            elif field == 'Мальчик родился':
                fields['sex'] = 'm'
            else:
                assert False
            field = 'Месяц рождения'
        else:
            assert field == fields_order[text_no]
        fields[field] = value
    data_with_fields.append(fields)
```

```python
counters = {field: Counter() for field in fields_order}
for field in fields_order:
    cnt = counters[field]
    for item in data_with_fields:
        cnt[item[field]] += 1
    print(field, len(cnt))
    print(counters[field].most_common(10))
    print()
```

```python
months = Counter()
years = Counter()
months_order = ['---', 'январе', 'феврале', 'марте', 'апреле', 'мае', 'июне', 'июле', 'августе', 'сентябре', 'октябре', 'ноябре', 'декабре']
for item in data_with_fields:
    month, year = item['Месяц рождения'].split(' ')
    item['year'] = int(year)
    item['month'] = months_order.index(month)
    months[item['month']] += 1
    years[item['year']] += 1
print(months.most_common())
print(years.most_common())
```
